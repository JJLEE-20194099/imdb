```python
import pandas as pd
import numpy as np
```

```python
ratings = pd.read_csv('data/datasets/rating/remaining_detail.csv', sep=',', encoding='latin-1')
ratings.head()
```

```python
ratings
```

```python
comment_and__rating_data = ratings[['content', 'rating']]
```

```python
comment_and__rating_data
```

```python
comment_and__rating_data = comment_and__rating_data.dropna(subset=['content', 'rating'])
```

```python
comment_and__rating_data = comment_and__rating_data.drop_duplicates(keep='first')
comment_and__rating_data
```

```python
import os
os.makedirs('data/datasets/comment_and_rating/', exist_ok = True)
output = os.path.join('data/datasets/comment_and_rating/', 'details.csv')
```

```python
print(output)
comment_and__rating_data.to_csv(output, sep=',', encoding='latin-1', index=False)
```
