```python
import pandas as pd
import numpy as np
```

```python
movie_details = pd.read_csv('data/datasets/movie/ml_details.csv', sep=',', encoding='latin-1')
movie_details.head()
```

```python
movie_details
```

```python
movie_details = movie_details.drop_duplicates(keep='first')
```

```python
movie_details
```

```python
movie_details['production companies']
```

```python
movie_details = movie_details.drop(['production companies'], axis=1)
movie_details
```

```python
movie_details['series'] = movie_details['series'].fillna('Not series')
```

```python
movie_details
```

```python
movie_details['release year'] = movie_details['release year'].fillna('undefined')
movie_details
```

```python
movie_details['official sites'] = movie_details['official sites'].fillna('undefined')
movie_details
```

```python
movie_details['certification'] = movie_details['certification'].fillna('others') 
movie_details
```

```python
movie_details['duration'] = movie_details['duration'].fillna('undefined')
movie_details
```

```python
movie_details['budget'] = movie_details['budget'].fillna('undefined')
movie_details
```

```python
import os

output = os.path.join('data/datasets/movie/', 'preprocessed_details.csv')
print(output)
movie_details.to_csv(output, sep=',', encoding='latin-1', index=False)
```

```python
content_and_average_rating_data = movie_details[['content', 'average rating']]
```

```python
content_and_average_rating_data
```

```python
content_and_average_rating_data = content_and_average_rating_data.dropna(subset=['content', 'average rating'])
```

```python
content_and_average_rating_data = content_and_average_rating_data.drop_duplicates(keep='first')
content_and_average_rating_data
```

```python
import os

output = os.path.join('data/datasets/content_and_rating/', 'details.csv')
print(output)
content_and_average_rating_data.to_csv(output, sep=',', encoding='latin-1', index=False)
```
