```python
import pandas as pd
import numpy as np
```

```python
ratings = pd.read_csv('./data/datasets/rating/details.csv', sep=',', encoding='latin-1')
ratings.head()
```

```python
ratings.info()
```

```python
ratings = ratings.dropna(subset=['rating'])
ratings
```

```python
duplicated_ratings = ratings[ratings.duplicated()]
duplicated_ratings
```

```python
ratings = ratings.drop_duplicates(keep='first')
```

```python
duplicated_ratings = ratings[ratings.duplicated()]
duplicated_ratings
```

```python
no_ratings_per_user = ratings.groupby('user id').count()['rating']
```

```python
no_ratings_per_user
```

```python
user_ids = ratings.groupby('user id').count()['rating'].index.tolist()
user_ids = np.array(user_ids)
```

```python
user_ids
```

```python
no_ratings_per_user = np.array(no_ratings_per_user)
user_id_idxes = np.argsort(no_ratings_per_user).tolist()
```

```python
user_ids = user_ids[user_id_idxes]
```

```python
for i in range(len(user_id_idxes[:1500])):
    idx = user_id_idxes[len(user_id_idxes) - i - 1]
    print(user_ids[idx], no_ratings_per_user[idx])
```

```python
eliminated_user_ids = user_ids[user_id_idxes[:-1500]]
len(eliminated_user_ids)
```

```python
interaction_users = pd.DataFrame(columns=['user index', 'user id', 'interaction count'])
interaction_users = interaction_users.reset_index(drop=True)
interaction_users['user index'] = range(len(user_ids))
interaction_users['user id'] = user_ids
interaction_users['interaction count'] = no_ratings_per_user[user_id_idxes]

```

```python
interaction_users
```

```python
movie_ids = open('./data/datasets/movie/ids.txt', 'r')
movie_ids = movie_ids.readlines()
movie_ids = set(movie_ids)
movie_ids = [movie_id.strip() for movie_id in movie_ids]
```

```python
movie_ids_df = pd.DataFrame(columns=['movie index', 'movie id'])
movie_ids_df['movie index'] = range(len(movie_ids))
movie_ids_df['movie id'] = movie_ids
```

```python
movie_ids_df.head()
```

```python
merged_movie_ids_and_ratings = pd.merge(movie_ids_df, ratings, how='inner', left_on='movie id', right_on='movie id')
merged_movie_ids_and_ratings
```

```python
merged_movie_ids_and_ratings_and_interaction_users = pd.merge(merged_movie_ids_and_ratings, interaction_users, how='inner', left_on='user id', right_on='user id')
```

```python
merged_movie_ids_and_ratings_and_interaction_users
```

```python
ratings = merged_movie_ids_and_ratings_and_interaction_users[['user index', 'movie index', 'rating', 'content', 'date', 'user id', 'movie id', 'interaction count']]
```

```python
ratings
```

```python
ratings.to_csv('./data/datasets/rating/preprocessed_detail.csv', sep=',', encoding='latin-1', index=False)
```
